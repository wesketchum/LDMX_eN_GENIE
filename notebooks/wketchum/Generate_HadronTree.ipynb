```python
gst_df = pd.DataFrame()
```

```python
for batch in uproot.iterate(["/Users/wketchum/Data/LDMX/eN_Ti_GENIE_v3_2_0/ldmx_eTi_4GeV_GEM21_11b_00/*gst.root:gst"],["pdgf","Ef","pxf","pyf","pzf"],library='pd'):
    gst_df = pd.concat([gst_df,batch])
```

```python
gst_df["pf"] = np.sqrt(gst_df["pxf"]*gst_df["pxf"]+gst_df["pyf"]*gst_df["pyf"]+gst_df["pzf"]*gst_df["pzf"])
```

```python
gst_df["mass2f"] = (gst_df["Ef"]*gst_df["Ef"])-(gst_df["pf"]*gst_df["pf"])
```

```python
gst_df["mass2f"] = np.where(gst_df["mass2f"] > 0, gst_df["mass2f"], 0)
```

```python
gst_df["massf"] = np.sqrt(gst_df["mass2f"])
```

```python
gst_df["kef"] = gst_df["Ef"]-gst_df["massf"]
```
